# 연습문제 2

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
```

```python
tips
```

```python
tip_pct_by_day_time_size = tips.pivot_table(values='tip_pct', index=['day', 'time', 'size'], columns='sex', aggfunc='mean')
```

```python
tip_pct_by_day_time_size
```

```python
tip_pct_by_day_time_size.idxmax()
```

```python
def peak_to_peak(x):
    return x.max() - x.min()
```

```python
tips.groupby(['sex', 'smoker'])[['tip']].agg(peak_to_peak)
```

```python
tips.groupby(['sex', 'smoker']).agg({'tip_pct': 'mean', 'total_bill': peak_to_peak})
```

```python
tips.pivot_table(values=['tip_pct', 'size'], index=['sex', 'day'], columns='smoker')
```

```python
tips.pivot_table('size', ['time', 'sex', 'smoker'], 'day', aggfunc='sum')
```

# 연습문제 3

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
pd.qcut?
```

```python
titanic['연령구간'] = pd.qcut(titanic.age, 3, labels=['청년', '중년', '노인'])
```

```python
titanic
```

```python
def pct_alive(x):
    return x.sum() / len(titanic)
```

```python
titanic.pivot_table(values='survived', index=['sex', '연령구간'], columns='class', aggfunc=pct_alive)
```

```python
titanic.groupby(['sex', '연령구간', 'class'])[['survived']].agg(pct_alive).unstack()
```
