# date_range

```python
pd.date_range('2024-1-22', '2024-4-1')
```

```python
pd.date_range('2024-1-22', periods=100)
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="B") # 주말이 아닌 평일
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="W") # 주(일요일)
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="W-MON") # 주(월요일)
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="MS") # 각 달의 첫날
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="M") # 각 달의 마지막날
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="BMS") # 주말이 아닌 평일 중에서 각 달의 첫날
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="BM") # 주말이 아닌 평일 중에서 각 달의 마지막 날
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="WOM-2THU") # 각 달의 두 번째 목요일
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="Q-JAN") # 각 분기의 첫 달의 마지막 날
```

```python
pd.date_range("2024-1-1", "2024-12-31", freq="Q-DEC") # 각 분기의 마지막 달의 마지막 날
```

```python
data_2014 = pd.read_csv('data/2014년 졸음운전 교통사고.csv', encoding='ANSI', index_col='구분')
data_2015 = pd.read_csv('data/2015년 졸음운전 교통사고.csv', encoding='ANSI', index_col='구분')
data_2016 = pd.read_csv('data/2016년 졸음운전 교통사고.csv', encoding='ANSI', index_col='구분')
data_2014.head()
```

```python
data_2014.info()
```

```python
pd.concat?
```

```python
data_total = pd.concat([data_2014, data_2015, data_2016], axis=0)
data_total
```

```python
data_total.info()
```

```python
data_total.index = pd.to_datetime(data_total.index, format='%Y년%m월')
data_total.info()
```

```python
data_total
```

```python
data_total['year'] = data_total.index.year
```

```python
data_total
```

```python
data_total['month'] = data_total.index.month
```

```python
data_total
```

```python
data_total['quarter'] = data_total.index.quarter
```

```python
data_total
```

```python
data_total.groupby(['quarter', 'year', 'month'])[['사망(명)']].mean()
```
