## X Only DataFrame

```python
DF = mini.iloc[:, 1:-8]
DF.head()
```

# K-means Clustering

## Modeling

```python
from sklearn.cluster import KMeans
```

```python
KMeans?
```

```python
kmeans_3 = KMeans(n_clusters=3,
                 init='k-means++',
                 max_iter=15,
                 random_state=2045)
kmeans_3.fit(DF)
```

## Clustering Results

```python
kmeans_3.n_iter_
```

```python
# 군집별 중심점
kmeans_3.cluster_centers_
```

```python
# 군집 결과 레이블
kmeans_3.labels_
```

```python
# 군집 중심까지의 제곱 거리의 합
kmeans_3.inertia_
```

## Scree Plot

K(1~9) 군집분석

```python
inertia = []

K = range(1,10)

for k in K:
    kmeanModel = KMeans(n_clusters= k)
    kmeanModel.fit(DF)
    inertia.append(kmeanModel.inertia_)
```

## 군집 중심까지의 제곱 거리의 합

```python
inertia
```

## Plot the elbow

```python
plt.figure(figsize=(9,7))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.title('Scree Plot with Kink')
plt.show()
```

# Visualization with PCA

## target 및 cluster 추가

```python
DF['cluster'] = kmeans_3.labels_
DF['target'] = mini.preferred_difficulty_level
```

```python
DF.head(3)
```

## 군집 결과 확인

```python
DF.groupby('target')['cluster'].value_counts() # low 1 medium 2 high 0
```

```python
mini.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(mini.iloc[:, 1:-8])
print(pca.explained_variance_ratio_) # 설명력
pca_transformed[:5]
```

## pca_x와 pca_y 추가

```python
DF['pca_x'] = pca_transformed[:, 0]
DF['pca_y'] = pca_transformed[:, 1]
```

```python
DF.head(5)
```

## 2차원 시각화

군집 값 0, 1, 2 인덱스 추출

```python
idx_0 = DF[DF['cluster'] == 0].index
idx_1 = DF[DF['cluster'] == 1].index
idx_2 = DF[DF['cluster'] == 2].index
```

```python
idx_0, idx_1, idx_2
```

0, 1, 2 인덱스 시각화

```python
plt.figure(figsize=(9,7))
plt.scatter(x=DF.loc[idx_0, 'pca_x'],
           y=DF.loc[idx_0, 'pca_y'],
           marker='o')
plt.scatter(x=DF.loc[idx_1, 'pca_x'],
           y=DF.loc[idx_1, 'pca_y'],
           marker='s')
plt.scatter(x=DF.loc[idx_2, 'pca_x'],
           y=DF.loc[idx_2, 'pca_y'],
           marker='^')
plt.xlabel('pca_x')
plt.ylabel('pca_y')
plt.title('Visualization by 2 PCA Components')
plt.show()
```
