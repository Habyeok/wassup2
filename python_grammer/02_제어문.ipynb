## while문 강제로 빠져나가기

```python
coffee = 10
money = 300

while money:
    print('돈을 받았으니 커피를 줍니다.')
    coffee = coffee - 1
    print('남은 커피는 %d잔 입니다.'% coffee)
    
    if coffee == 0:
        print('커피가 다 떨어졌습니다. 판매를 중지합니다.')
        break
```

총 5명의 학생이 시험을 보았는데 시험 점수가 60점 이상이면 합격이고 그렇지 않으면 불합격이다. 합격인지, 불합격인지 결과를 보여 주시오.

```python
scores = [90, 25, 67, 45, 80]
```

```python
number = 0

for score in scores:
    number += 1
    
    if score >= 60:
        print(f'{number}번 학생의 점수는 {score}점이므로 합격입니다.')
    else:
        print(f'{number}번 학생의 점수는 {score}점이므로 불합격입니다.')
```

## continue문

```python
marks = [90, 25, 67, 45, 80]

number = 0 
for mark in marks: 
    number = number +1 
    if mark < 60:
        continue 
    print("%d번 학생 축하합니다. 합격입니다. " % number)
```
